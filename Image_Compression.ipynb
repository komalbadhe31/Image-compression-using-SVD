```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os
plt.rcParams['figure.figsize']=[12,6]

A = imread('img.jpeg')
img = plt.imshow(A)
plt.axis('off')
plt.show()
```

```python
X = np.mean(A,-1)

img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
from numpy.core.numeric import full
U, S, VT =  np.linalg.svd(X,full_matrices=False)

# corresposnding to non zero eigen values
S = np.diag(S)  # just extract the diagonal elements, since non daigonal are ZERO

j=0 # just a numeric counter
for r in ( 5, 20, 100): # r is the no of components taken
    # construct approximate image
    Xapprox = U[ : , :r] @ S[0:r, :r] @ VT[ : r, : ]
    # we take all the rows of U and first r columns of U
    # we take only r rows and r columns of S
    # we take r rows of VT and all columns
    plt.figure(j+1)
    j += 1
    img = plt.imshow(Xapprox)
    plt.set_cmap('gray')
    #plt.axis('off')
    plt.title('r = '+str(r))
    plt.show()
```

```python
plt.figure(1)

plt.semilogy(np.diag(S))
plt.title('Singular values')
plt.xlabel('j',fontsize =20)
plt.ylabel('$log(\sigma_j)$',fontsize =20)
plt.show()
```
